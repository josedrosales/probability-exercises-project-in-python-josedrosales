## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import numpy as np

iteraciones = 1000
contador = 0

for i in range(iteraciones):

    dado1 = np.random.randint(1, 7)
    dado2 = np.random.randint(1, 7)

    suma = dado1 + dado2

    if (suma % 2 == 0) or (suma > 7):
        
        contador += 1

probabilidad = contador / iteraciones
print(f"La probabilidad es: {probabilidad}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"
        
# TODO
iteraciones = 1000
suma_prob_1 = 0
suma_prob_2 = 0

for i in range(iteraciones):
    
    b1 = np.random.randint(60)
    b2 = np.random.randint(60)
    b3 = np.random.randint(60)
    b4 = np.random.randint(60)
    b5 = np.random.randint(60)
    bolas = [b1,b2,b3,b4,b5]    

    colores = [ball_box[key] for key in bolas]

    conteo_red = colores.count('Red')
    conteo_white = colores.count('White')
    conteo_green = colores.count('Green')

    if (conteo_white == 3) and (conteo_red == 2):

        suma_prob_1 += 1

    elif (conteo_white == 5) or (conteo_red == 5) or (conteo_green ==5):

        suma_prob_2 += 1

prob1 = suma_prob_1 / iteraciones
prob2 = suma_prob_2 / iteraciones

print(f"La probabilidad de obtener 3 pelotas blancas y dos rojas es: {prob1}")
print()
print(f"La probabilidad de obtener todas las pelotas del mismo color es: {prob2}")
```
